```python
import pandas as pd
import seaborn as sns

pd.set_option("display.max_rows", 2000)
```

```python
data = pd.read_csv("final_data_27_02.csv")
```

```python
data.columns
```

```python
rel_data = data[['1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']]
```

```python
rel_data.head()
```

```python
def select_year(data, year):
    data_of_year = data[data[year].notnull()]
    means_of_year = data_of_year.mean(axis=0)
    
    return means_of_year[["danceability", "energy", "loudness", "speechiness", "tempo", "valence"]]
    

fot = pd.DataFrame(columns=["danceability", "energy", "loudness", "speechiness", "tempo", "valence"])

years = ['1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

for year in years:
    fot.loc[year] = select_year(rel_data, year)

fot.to_csv("fot.csv")
```
