```python
import pandas as pd
import pickle
```

```python
data = pickle.load(open("final_data_27_02.pickle", "rb"))
```

```python
freqs = {}
for genres in data["genre"]:
    for genre in genres:
        if genre in freqs:
            freqs[genre] += 1
        else:
            freqs[genre] = 1
```

```python
len(freqs)
```

```python
highest_genres = []
for genres in data["genre"]:
    count = 0
    highest = None
    for genre in genres:
        if freqs[genre] > count:
            highest = genre
    highest_genres.append(genre)
```

```python
unique_highest_genres = list(set(highest_genres))
```

```python
len(unique_highest_genres)
```

```python
freqs_highest = {}
for genre in highest_genres:
    if genre in freqs_highest:
        freqs_highest[genre] += 1
    else:
        freqs_highest[genre] = 1
```
