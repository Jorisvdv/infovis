```python
import pandas as pd
import pickle
```

```python
data = pickle.load(open("final_data_27_02.pickle", "rb"))
```

```python
freqs = {}
for genres in data["genre"]:
#     print(genres)
#     print(type(genres))
    for genre in genres:
#         print(genre)
        if genre in freqs:
            freqs[genre] += 1
        else:
            freqs[genre] = 1
```
